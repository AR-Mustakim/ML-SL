```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
```

# load data

```python
dfKG = pd.read_csv('tweets_data/1_KG_4-26.csv')
dfKN = pd.read_csv('tweets_data/3_KN_4-29.csv')
dfSH = pd.read_csv('tweets_data/4_SH_4-29.csv')
dfCM = pd.read_csv('tweets_data/5_CM_4-29.csv')
dfJG = pd.read_csv('tweets_data/7_JG_4-29.csv')
dfUPSR = pd.read_csv('tweets_data/8_UPSR_5-5.csv')
dfARN = pd.read_csv('tweets_data/9_ARN_5-5.csv')
```

## DFKG

```python
stop_words = ["t", "nan", "s","di","the","this","i","in","of","or","is","if","it","at",'hashtag','ke','co','yang','u','https']
```

```python
#visualize Word cloud - plotting
plt.figure(figsize = (20,10))
allWords =' '.join([twts for twts in dfKG['Raw']])
wordCloud = WordCloud(width = 900, height=700, random_state =21, max_font_size=110, max_words=70,  stopwords = stop_words).generate(allWords)

plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize = (20,10))
allWords =' '.join([str(twts) for twts in dfKG['CD-RE']])
wordCloud = WordCloud(width = 900, height=700, random_state =21, max_font_size=110, stopwords=stop_words, max_words=70).generate(allWords)

plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
